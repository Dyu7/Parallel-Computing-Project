```python
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
import numpy as np
```

```python
#MNIST dataset
#gpu data points
mnist_data_points_gpu = []
data_point_1 = {'Number_of_Samples':50,'Iterations':100,'Accuracy':0.42,'Execution Time':3.749}
mnist_data_points_gpu.append(data_point_1)
data_point_2 = {'Number_of_Samples':50,'Iterations':1000,'Accuracy':0.92,'Execution Time':12.483}
mnist_data_points_gpu.append(data_point_2)
data_point_3 = {'Number_of_Samples':100,'Iterations':1000,'Accuracy':0.89,'Execution Time':17.526}
mnist_data_points_gpu.append(data_point_3)
data_point_4 = {'Number_of_Samples':100,'Iterations':10000,'Accuracy':0.90,'Execution Time':143.580}
mnist_data_points_gpu.append(data_point_4)
data_point_5 = {'Number_of_Samples':200,'Iterations':1000,'Accuracy':0.77,'Execution Time':24.538}
mnist_data_points_gpu.append(data_point_5)
data_point_6 = {'Number_of_Samples':200,'Iterations':10000,'Accuracy':0.90,'Execution Time':225.242}
mnist_data_points_gpu.append(data_point_6)
```

```python
#cpu data points
mnist_data_points_cpu = []
data_point_1 = {'Number_of_Samples':50,'Iterations':100,'Accuracy':0.38,'Execution Time':14.543}
mnist_data_points_cpu.append(data_point_1)
data_point_2 = {'Number_of_Samples':50,'Iterations':1000,'Accuracy':0.98,'Execution Time':129.912}
mnist_data_points_cpu.append(data_point_2)
data_point_3 = {'Number_of_Samples':100,'Iterations':1000,'Accuracy':0.81,'Execution Time':246.142}
mnist_data_points_cpu.append(data_point_3)
data_point_4 = {'Number_of_Samples':100,'Iterations':10000,'Accuracy':0.96,'Execution Time':2284.952}
mnist_data_points_cpu.append(data_point_4)
data_point_5 = {'Number_of_Samples':200,'Iterations':1000,'Accuracy':0.73,'Execution Time':512.873}
mnist_data_points_cpu.append(data_point_5)
data_point_6 = {'Number_of_Samples':200,'Iterations':10000,'Accuracy':0.92,'Execution Time':4789.304}
mnist_data_points_cpu.append(data_point_6)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel('Number of Samples')
ax.set_ylabel('Iterations')
ax.set_zlabel('Execution Time(s)')
ax.set_title('GPU Execution Times')

X, Y, Z = [],[],[]
for i in mnist_data_points_gpu:
    X.append(i['Number_of_Samples'])
    Y.append(i['Iterations'])
    Z.append(i['Execution Time'])
X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)
ax.plot3D(X, Y, Z)

plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel('Number of Samples')
ax.set_ylabel('Iterations')
ax.set_zlabel('Execution Time(s)')
ax.set_title('CPU Execution Times')

X, Y, Z = [],[],[]
for i in mnist_data_points_cpu:
    X.append(i['Number_of_Samples'])
    Y.append(i['Iterations'])
    Z.append(i['Execution Time'])
X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)
ax.plot3D(X, Y, Z)

plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel('Number of Samples')
ax.set_ylabel('Iterations')
ax.set_zlabel('Speedup')
ax.set_title('GPU Speedup')

X, Y, Z = [],[],[]
for i in range(len(mnist_data_points_cpu)):
    X.append(mnist_data_points_cpu[i]['Number_of_Samples'])
    Y.append(mnist_data_points_cpu[i]['Iterations'])
    Z.append(mnist_data_points_cpu[i]['Execution Time'] / 
             mnist_data_points_gpu[i]['Execution Time']
            )
X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)
ax.plot3D(X, Y, Z)

plt.show()
```

```python
fig,ax = plt.subplots()
ax.set_xlabel('Number of Samples x Iterations')
ax.set_ylabel('Execution Time(s)')
ax.set_title('GPU Execution Times')

X, Y = [],[]
for i in mnist_data_points_gpu:
    X.append(i['Number_of_Samples']*i['Iterations'])
    Y.append(i['Execution Time'])
#sorting inputs
Y = [y for _,y in sorted(zip(X,Y))]
X = sorted(X)
X = np.array(X)
Y = np.array(Y)
ax.plot(X, Y)
```
